```python
import pandas as pd
import requests
import json
```

```python
symbol = '2330'
```

### TWSE

```python
twse_api_url = 'https://openapi.twse.com.tw/v1/exchangeReport/STOCK_DAY_ALL'
res = requests.get(twse_api_url)
data = json.loads(res.text)

twse_df = pd.DataFrame(data)

twse_df
```

```python
twse_df[twse_df['Code'] == symbol]
```
